```python
vg = pd.concat([df1,df2],axis=1)
```

```python
vg.dropna(inplace=True)
```

```python
vg.drop_duplicates(inplace=True)
```

```python
vg.Year = vg.Year.astype(int)
```

```python
vg['Total_Sales'] = vg.NA_Sales + vg.EU_Sales + vg.JP_Sales
```

```python
plt.figure(figsize=(12,3))
plt.bar(vg.Genre, vg.NA_Sales)
plt.ylabel('NA Sales')
plt.xlabel('Genre')
plt.title('Genre wise NA Sales')
plt.show()
```

```python
global_mean = vg['Global_Sales'].mean()
north_america_mean = vg['NA_Sales'].mean()
europe_mean = vg['EU_Sales'].mean()
japan_mean = vg['JP_Sales'].mean()
others_mean = vg['Other_Sales'].mean()
labels = ['Global Sales','NA Sales','EU Sales',"JP Sales","Others Sales"]
x_pos = [0, 1,2,3,4]
heights = [global_mean,north_america_mean,europe_mean,japan_mean,others_mean]
plt.bar(x_pos, heights, align='center')
plt.xticks(x_pos, labels)
plt.ylabel('Mean Sales')
plt.title('Mean Sales by Region')
plt.show()
```

```python
def group_and_sum(vg, column):
    grouped = vg.groupby(column)['Global_Sales'].sum()
    return grouped
```

```python
grouped_by_category = group_and_sum(vg,'Genre')
print('Grouped by Category:')
print(grouped_by_category)
```

```python
sales_by_publisher = vg.groupby(['Publisher','Year'])[['Total_Sales']].sum()
```

```python
plt.figure(figsize=(12,8))
for publisher in sales_by_publisher.index.get_level_values('Publisher').unique():
    data = sales_by_publisher.loc[publisher]
    plt.plot(data.index, data.values)
plt.legend()
plt.title('Total Sales by Publisher')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()         #989 Sports
```
